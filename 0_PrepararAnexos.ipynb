```python
import pandas as pd
```

```python
df_cod_ens = pd.read_csv("https://raw.githubusercontent.com/rulehial/Analisis-Educacion/main/Datos/AnexoCodigoEnsenanza.csv", sep=";")
df_cod_ens
```

```python
df_cod_ens.info()
```

```python
def agregaNivel(fila):
  if fila["COD_ENSE"] == 10:
    val = "Parvularia"
  elif 110 <= fila["COD_ENSE"] <= 167:
    val = "Básica"
  elif 211 <= fila["COD_ENSE"] <= 299:
    val = "Especial"
  elif fila["COD_ENSE"] >= 310:
     val = "Media"
  else:
    val = "Otra"
  return val
```

```python
# df_cod_ens["nivel"] =  ["Parvularia" if s == 10 else "Otro" for s in df_cod_ens['COD_ENSE']]
df_cod_ens["nivel"] = df_cod_ens.apply(agregaNivel, axis=1)
df_cod_ens
```

```python
df_cod_ens.to_csv('AnexoCodigoEnsenanzaNivel.csv', index=False)
```
